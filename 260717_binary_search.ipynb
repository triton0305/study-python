```python
arr = list(map(int, input("숫자를 공백으로 구분해 입력해주세요. >> ").split()))
search = int(input("찾을 값을 입력하세요 >> "))

if len(arr) == 0:
    print("해당 값은 존재하지 않습니다.")
else:
    lowIndex = 0
    highIndex = len(arr)-1
    arr.sort()
    while True:
        midIndex = (lowIndex + highIndex) // 2
        if search < arr[midIndex]:
            if midIndex == lowIndex:
                print("해당 값은 존재하지 않습니다.")
                break
            highIndex = midIndex-1
        elif search > arr[midIndex]:
            if highIndex == midIndex:
                print("해당 값은 존재하지 않습니다.")
                break
            lowIndex = midIndex+1
        else:
            print(f"해당값은 index: {midIndex}에 있습니다.")
            break
```

```python
## 프로그래머스 야근지수
## 시간 초과 코드
## import heapq
## heapq.heapify(works), heapq.heappush(works, heapq.heappop(works) + 1), heapq.heapreplace(works, works[0] + 1)  

def solution(n, works):
    answer = 0
    sum = 0
    for i in range(0,len(works)):
        sum+=works[i]
    if sum > n:
        while (n > 0) :
            works.sort()
            works.reverse()
            works[0] -= 1
            n -= 1
        for j in range (0, len(works)):
            answer += pow(works[j], 2)
    return answer
```

```python
## 트리, 부모, 자식, 루트, 리프
## heapq : 밑에서 위로, 숫자 작은 자식과 부모 비교

from heapq import heappop, heappush, heapify
def solution(n, works):
    answer = 0
    total = 0
    for i in range(0,len(works)):
        total+=works[i]
        works[i] *= -1
    heapify(works)
    if total > n:
        while (n > 0) :
            heappush(works, heappop(works) + 1)
            n -= 1
        for j in range (0, len(works)):
            answer += pow(works[j], 2)
    return answer
```
